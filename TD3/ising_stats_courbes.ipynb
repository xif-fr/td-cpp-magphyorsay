```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize=(8,8), sharex=True)

for i in range(5):
    data = np.loadtxt(f"ising_magntisation_nergie_txaccept.{i}.txt")
    steps = data[::30,0]
    magnet = data[::30,1]
    nergie = data[::30,2]
    txaccept = data[::30,3]
    ax1.plot(steps, txaccept, alpha=0.7)
    ax2.plot(steps, magnet, alpha=0.7)
    ax3.plot(steps, nergie, alpha=0.7)
    
ax1.set_ylim(0,1)
ax1.set_ylabel("Taux d'acceptation")

ax2.set_ylim(-1,1)
ax2.set_ylabel("Magntisation")

ax3.set_ylabel("nergie interne")

ax3.set_xlabel("Nombre de pas de l'algorithme")
ax3.set_xlim(-10000000,1000000000)

plt.tight_layout()
plt.savefig("ising_stats_courbes_T1_B0_nx200_ny100.pdf", transparent=True)
```
