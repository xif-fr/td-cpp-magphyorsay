```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
a = 0.01
Iz = np.loadtxt("newton_bassins_converg.txt", dtype=int)
Iz
```

```python
Iz = Iz.transpose()[::-1,:]
```

```python
RGB = np.zeros((Iz.shape[0],Iz.shape[1],3))

RGB[ Iz == 0, : ] = [1,0,0]
RGB[ Iz == 1, : ] = [0.7,0,0.7]
RGB[ Iz == 2, : ] = [0,0,1]
```

```python
traj = np.loadtxt("newton_traj.txt")
```

```python
plt.figure(figsize=(10,10))
plt.imshow(RGB, extent=[-a,a,-a,a], interpolation='none')
plt.xlabel(r"$\mathfrak{Re}(z_0)$")
plt.ylabel(r"$\mathfrak{Im}(z_0)$")
# plt.plot(traj[:,0], traj[:,1], '--.', color='black', lw=1.2, alpha=0.3)
plt.savefig("newton_bassins_converg.a001.pdf", bbox_inches='tight', transparent=True)
```
